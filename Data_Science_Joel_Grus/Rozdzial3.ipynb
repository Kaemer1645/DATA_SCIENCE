```python
#praca z matplotlib

from matplotlib import pyplot as plt
from random import randrange
years = [date for date in range(1950,2020, 10)]
gdp = [randrange(300,15000) for _ in range(7)]
```

```python
#prosty wykres liniowy

plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')
plt.title("Nominalny PKB")
plt.ylabel("Mld dol.")
plt.show()
```

```python
#wykres słupkowy

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(range(len(movies)), num_oscars)
plt.ylabel("Liczba nagrod")
plt.title("Moje ulubione filmy")

plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
from collections import Counter

grades = [randrange(0,100) for grade in range(13)]

#podziel stopnie na decyle, ale tak, zeby wartosc 100 znalazla sie w jednym przedziale z 91 i 95
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()],
       histogram.values(),
       10)

plt.axis([-5, 105, 0 , 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decyl")
plt.ylabel("Liczba studentow")
plt.title("Rozklad ocen z pierwszego egzaminu")
plt.show()
```

```python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("ile razy uslyszalem fraze nauka o danych")
plt.ticklabel_format(useOffset = False)
plt.axis([2016.5, 2018.5, 499,506])
plt.title("Patrz jaki ogromny wzrost!")
plt.show()


plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("ile razy uslyszalem fraze nauka o danych")
plt.ticklabel_format(useOffset = False)
plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Niewielki wzrost")
plt.show()
```

```python
#wykresy liniowe

variance = [2 ** _ for _ in range(9)]
bias_squared = variance
bias_squared = bias_squared[::-1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label = 'variance')
plt.plot(xs, bias_squared, 'r-.', label = 'bias ** 2')
plt.plot(xs, total_error, 'b:', label = 'total_error')

plt.legend(loc=9) #loc=9 daje legende na srodku u gory wykresu
plt.xlabel("Stopien skomplikowania modelu")
plt.title("Kompromis pomiędzy wartoscia progowa i zaleznoscia modelu")
plt.show()
```

```python
#wykresy punktowe
import string


friends = [70, 65, 72, 63, 71, 64, 60 ,64 ,67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = [string.ascii_lowercase[_] for _ in range(9)]
plt.scatter(friends, minutes)

#nadanie etykiet wszystkim punktom
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, xy=(friend_count, minute_count),
                xytext = (5, -5),
                textcoords='offset points')
    
    
plt.title("Czas spedzony na stronie a liczba znajomych")
plt.xlabel("Liczba znajomych")
plt.ylabel("Dzienny czas spedzony na stronie (minuty)")
plt.show()
```
