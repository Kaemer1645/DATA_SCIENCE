```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('World_Bank_CO2_cleaned.xlsx')
df.head()
```

```python
df = df[~df['CO2 (kt)'].isnull() & ~df['CO2 Per Capita (metric tons)'].isnull() & (df['Year'] > 1980)]

df.head()
```

```python
df_pl = df[df['Country Name'] == 'Poland']
df_pl
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(data=df[(df['Country Name'] == 'Poland') | (df['Country Name'] == 'China') | (df['Country Name'] == 'Germany')],
           x='Year',
           y='CO2 (kt)',
           aspect=2.5,
            kind= 'line',
            hue='Country Name')

plt.show()


#aspect is a width of graph
```

```python
sns.relplot(data=df[(df['Country Name'] == 'Poland') | (df['Country Name'] == 'Germany')],
           x='Year',
           y='CO2 Per Capita (metric tons)',
           aspect=2.5,
            kind= 'scatter',
            col='Country Name')

plt.show()

#col - places graph in different columns
```

```python
sns.relplot(data=tips,
           x='total_bill',
           y='tip',
           aspect=2.5,
            size='size',
            kind= 'scatter',
            hue='smoker')

plt.show()
```

```python
sns.catplot(data=df[df['Year']==2010],
           x='Region',
           y='CO2 Per Capita (metric tons)',
           aspect=2.5)

plt.show()
```

```python
sns.catplot(data=tips,
           x='day',
           y='tip',
           aspect=2.5,
            kind='box',
           hue='sex')

plt.show()
```

```python
sns.lmplot(data=df[(df['Country Name'] == 'Poland') | (df['Country Name'] == 'Germany')],
           x='Year',
           y='CO2 Per Capita (metric tons)',
           aspect=2.5,
            hue='Country Name')

plt.show()
```

```python
sns.distplot(tips['tip'],bins=10)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.distplot(tips[tips['day']=='Sun']['tip'], bins=10, kde_kws={"label": "Sunday"})
sns.distplot(tips[tips['day']=='Fri']['tip'], bins=10, kde_kws={"label": "Friday"})
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.relplot(data=df_pl,x="Year",y="CO2 (kt)",aspect=2)
plt.show()
```

```python
sns.set_style("whitegrid")

sns.relplot(data=df_pl,x="Year",y="CO2 (kt)",aspect=2)
plt.show()
```

```python
sns.set_style("dark")
sns.relplot(data=df_pl,x="Year",y="CO2 (kt)",aspect=2)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_context("poster")
sns.relplot(data=df_pl,x="Year",y="CO2 (kt)",aspect=2, kind='scatter')
plt.show()
```

```python
sns.set_context("paper")
sns.relplot(data=df_pl,x="Year",y="CO2 (kt)",aspect=2, kind='line')
plt.title("Emission of CO2 - Poland vs Germany vs Italy")
plt.xlabel("Years")
plt.ylabel("CO2")
plt.show()
```
