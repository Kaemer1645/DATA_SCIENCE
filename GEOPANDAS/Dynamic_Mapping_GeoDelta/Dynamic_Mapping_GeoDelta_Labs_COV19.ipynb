```python
import pandas as pd
import geopandas as gpd

import os
```

```python
data_confirmed = pd.read_html('https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
for df_confirmed in data_confirmed:
    pass
df_confirmed = df_confirmed
df_confirmed.head()
```

```python
df_confirmed = df_confirmed.groupby('Country/Region').sum()
```

```python
df_confirmed = df_confirmed.drop(columns = ['Lat','Long','Unnamed: 0'])
```

```python
world.replace('Iran (Islamic Republic of)','Iran',inplace=True)
world.replace('Brunei Darussalam','Brunei',inplace=True)
world.replace('Cape Verde','Cabo Verde',inplace=True)
world.replace('Congo','Congo (Brazzaville)',inplace=True)
world.replace('Congo','Congo (Kinshasa)',inplace=True)
world.replace('Czech Republic','Czechia',inplace=True)
#world.replace('','Eswatini',inplace=True)
#world.replace('','Holy See',inplace=True)
world.replace('Korea, Republic of','Korea, South',inplace=True)
#world.replace('','Kosovo',inplace=True)
world.replace('Lao People\'s Democratic Republic','Laos',inplace=True)
world.replace('Libyan Arab Jamahiriya','Libya',inplace=True)
#world.replace('','MS Zaandam',inplace=True)
world.replace('Republic of Moldova','Moldova',inplace=True)
#world.replace('','North Macedonia',inplace=True)
world.replace('Sudan','South Sudan',inplace=True)
world.replace('Syrian Arab Republic','Syria',inplace=True)
world.replace('Taiwan','Taiwan*',inplace=True)
#world.replace('','Tanzania',inplace=True)
world.replace('United States','US',inplace=True)
world.replace('Viet Nam','Vietnam',inplace=True)
#world.replace('','West Bank and Gaza',inplace=True)
```

```python
merge_files = world.join(df_confirmed, on= 'NAME', how='right')

merge_files
```

```python
#CREATE GIF - import libs
import PIL
import io

collect_img_list = []
```

```python
%%capture
list_of_dates = merge_files.columns.tolist()

for data in list_of_dates[2:]:
    plotter = merge_files.plot(column=str(data), cmap='OrRd',figsize=(10,10),legend = True,
                               scheme = 'user_defined', edgecolor="black",
                               classification_kwds = {'bins' : [100,10000,100000,1000000,5000000]})
                              #legend_kwds={'label': "Covid-19 Confirmed Cases",'orientation': "horizontal"})
    #add title to the map
    plotter.set_title("Covid-19 Confirmed Cases "+str(data), fontsize=25)
    #remove axis
    plotter.set_axis_off()
    #move legend
    plotter.get_legend().set_bbox_to_anchor((0.2,0.1))
    
    
    
    
    #GIF prepare
    img = plotter.get_figure()
    
    f = io.BytesIO()
    img.savefig(f, format = 'png', bbox_inches = 'tight')
    f.seek(0)
    collect_img_list.append(PIL.Image.open(f))
    
    

    




```

```python
#create animation 
collect_img_list[0].save('Dynamic_COV-19_MAP.gif', format = 'GIF',
                        append_images = collect_img_list[1:],
                        save_all = True, duration = 150,
                        loop = 0)
f.close()
```
