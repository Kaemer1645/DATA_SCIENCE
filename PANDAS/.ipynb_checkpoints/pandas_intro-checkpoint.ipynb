```python
df_pop = df[['Country','Population','Climate','Phones per 1000']].copy()
df_pop
```

```python
df_pop['Population'] /= 1000000
df_pop['Phones per 1000'] = df_pop['Phones per 1000'].replace(np.nan, 0)
df_pop
df_pop['Phones per 1000'] = df_pop['Phones per 1000'].str.replace(',', '.')
df_pop['Phones per 1000'] = pd.to_numeric(df_pop['Phones per 1000'], downcast="float")
df_pop
#df['Phones per 1000'] = df['Phones per 1000'].astype(float)
```

```python
new_col = []
for val in df_pop['Phones per 1000']:
    if val <= 100:
        new_col.append(True)
    elif val > 100:
        new_col.append(False)
    else:
        new_col.append(None)
print(new_col)
print(len(new_col))
```

```python
df_pop['New Column'] = new_col
df_pop[:10]
```
