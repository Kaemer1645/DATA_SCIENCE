```python
df_pop = df[['Country','Region','Population','Climate','Phones per 1000']].copy()
df_pop
```

```python
df_pop['Population'] /= 1000000
df_pop['Phones per 1000'] = df_pop['Phones per 1000'].replace(np.nan, 0)
df_pop
df_pop['Phones per 1000'] = df_pop['Phones per 1000'].str.replace(',', '.')
df_pop['Phones per 1000'] = pd.to_numeric(df_pop['Phones per 1000'], downcast="float")
df_pop
#df['Phones per 1000'] = df['Phones per 1000'].astype(float)
```

```python
new_col = []
for val in df_pop['Phones per 1000']:
    if val <= 100:
        new_col.append(True)
    elif val > 100:
        new_col.append(False)
    else:
        new_col.append(None)
```

```python
df_pop['New Column'] = new_col
df_pop.head()
```

```python
for index, row in df_pop.iterrows():
    if row['Population'] > 100:
        df_pop.loc[index,'Size'] = 'Big'
        print(row['Country'] , df_pop.loc[index, 'Size'])
```

```python
df_pop
```

```python
for row in df_pop.itertuples():
    if row.Population > 200:
        print(row.Index, row.Country, row.Population)
```

```python
print(df_pop.Population.sum())
print(df_pop.Population.min())
print(df_pop.Population.max())
print(df_pop.Population.mean())
```

```python
print(df_pop.groupby('Region')['Population'].sum())
print(df_pop.groupby('Region')['Population'].min())
print(df_pop.groupby('Region')['Population'].max())
print(df_pop.groupby('Region')['Population'].mean())
```
